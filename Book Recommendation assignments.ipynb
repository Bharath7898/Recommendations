```python
bdf = book_df.drop('Unnamed: 0',axis=1)
bdf1 = bdf.rename({'User.ID': 'User_ID'}, axis=1)
bdf2 = bdf1.rename({'Book.Title': 'Book_Title'}, axis=1)
bdf3 = bdf2.rename({'Book.Rating': 'Book_Rating'}, axis=1)
bdf3
```

```python
#Count of duplicated rows
bdf3[bdf3.duplicated()].shape
```

```python
bdf4=bdf3.drop_duplicates()
bdf4
```

```python
#number of unique users in the dataset
len(bdf4.User_ID.unique())
```

```python
len(bdf4.Book_Title.unique())
```

```python
len(bdf4.Book_Rating.unique())
```

```python
user_books_df = pd.pivot_table(bdf4,index='User_ID',
                                 columns='Book_Title',
                                 values='Book_Rating',aggfunc = np.sum)

user_books_df
```

```python
#Impute those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
user_books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_books_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = bdf4.User_ID.unique()
user_sim_df.columns = bdf4.User_ID.unique()
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```
